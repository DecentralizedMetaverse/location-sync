## 移動シミュレーション

## パケットサイズ

```python
size_head = 0 # byte
size_data = 14 # byte

max_visible = 10
max_visible_half = int(max_visible/2)

pps = 30

#  --------------------------------------------
n={}
n[0] = max_visible * (size_head + size_data)
n[1] = max_visible * (size_head + max_visible * size_data)
n[2] = max_visible_half * (size_head + (size_data) + (max_visible * size_data))
n[3] = max_visible_half * (size_head + (size_data) + (max_visible * size_data))
n[4] = max_visible_half * (size_head + (size_data))
n[5] = max_visible_half * (size_head + (size_data))

print("[node一つ当たりのパケット量]")
print(f"C/S \t: {n[0]:>16,} bytes")
print(f"C/S \t: {n[1]:>16,} bytes")
print(f"p2p\t: {n[2]:>16,} bytes")
print(f"p2p\t: {n[3]:>16,} bytes")
print(f"p2p\t: {n[4]:>16,} bytes")
print(f"p2p\t: {n[5]:>16,} bytes")
print()
```

```python
size_head = 44 # byte
size_data = 14 # byte

num_node = 1000
max_visible = 9
max_visible_half = int(max_visible/2)

pps = 30

#  --------------------------------------------
n={}
n[0] = size_head + size_data
n[1] = max_visible * (size_head + max_visible * size_data)
n[2] = max_visible_half * (size_head + (size_data))
n[3] = max_visible_half * (size_head + (size_data) + (max_visible_half * size_data))

print("[node一つ当たりのパケット量]")
print(f"C/S 上\t\t\t: {n[0]:>16,} bytes")
print(f"C/S 下\t\t\t: {n[1]:>16,} bytes")
print(f"p2p-mesh\t\t: {n[2]:>16,} bytes")
print(f"p2p-mesh(有他node情報)\t: {n[3]:>16,} bytes")
print()

r={}
r[0] = n[0]*num_node
r[1] = n[1]*num_node
r[2] = n[2]*num_node
r[3] = n[3]*num_node

print("[全体パケット総量]")
print(f"C/S 上\t\t\t: {r[0]:>16,} bytes")
print(f"C/S 下\t\t\t: {r[1]:>16,} bytes")
print(f"p2p-mesh\t\t: {r[2]:>16,} bytes")
print(f"p2p-mesh(有他node情報)\t: {r[3]:>16,} bytes")
print()




```
